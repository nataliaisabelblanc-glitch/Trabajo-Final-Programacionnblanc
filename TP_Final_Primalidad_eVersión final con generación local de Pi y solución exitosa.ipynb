```python
import math

def es_primo(numero):
    """
    Evalúa si un número entero es primo utilizando el criterio
    de optimización por raíz cuadrada (búsqueda acotada).
    Devuelve True si es primo, o False si es compuesto.
    """
    # Casos base: menores o iguales a 1 no son primos
    if numero <= 1:
        return False

    # El 2 es el único primo par
    if numero == 2:
        return True

    # Descartamos todos los números pares mayores a 2 al principio
    if numero % 2 == 0:
        return False

    # Calculamos la cota superior matemáticamente: parte entera de la raíz cuadrada
    limite = int(math.isqrt(numero))

    # Iteramos solo por los números impares desde 3 hasta la raíz cuadrada inclusive
    for divisor in range(3, limite + 1, 2):
        if numero % divisor == 0:
            return False  # Se encontró un divisor, es compuesto

    return True  # No se encontraron divisores, es primo

# =============================================================================
# BLOQUE DE PRUEBA (TESTING)
# =============================================================================
# Probamos la función con diferentes casos para verificar su consistencia lógica
print("--- VERIFICACIÓN DEL MOTOR DE PRIMALIDAD ---")
print(f"¿El número 7 es primo? {es_primo(7)}")     # Debería dar True
print(f"¿El número 15 es primo? {es_primo(15)}")   # Debería dar False
print(f"¿El número 2 es primo? {es_primo(2)}")     # Debería dar True
print(f"¿El número 1 es primo? {es_primo(1)}")     # Debería dar False
```

```python
# Definimos una cadena con los primeros decimales de e (sin la coma inicial)
# e = 2.7182818284590452353...
cadena_e = "27182818284590452353"

def buscar_primer_primo(cadena, tamano_bloque):
    """
    Recorre la cadena emulando una ventana deslizante de ancho 'tamano_bloque'.
    Convierte cada bloque a entero y lo evalúa con la función es_primo.
    """
    # Calculamos cuántas posiciones máximas se puede mover la ventana
    posiciones_maximas = len(cadena) - tamano_bloque + 1

    # Bucle iterativo que mueve el índice de inicio de la ventana
    for i in range(posiciones_maximas):
        # 1. REBANADO (Slicing): Extraemos el bloque de texto
        bloque_texto = cadena[i : i + tamano_bloque]

        # 2. CONVERSIÓN (Casting): Transformamos el texto a número entero
        numero_evaluar = int(bloque_texto)

        # 3. EVALUACIÓN: Le pasamos el número al motor que creamos en la celda anterior
        if es_primo(numero_evaluar):
            print(f"¡ÉXITO ENCONTRADO!")
            print(f"El primer primo de {tamano_bloque} dígitos es: {numero_evaluar}")
            print(f"Se encontró en la posición de inicio (índice): {i}")
            return numero_evaluar  # Cortamos la función inmediatamente

    print("Se recorrió toda la cadena y no se halló ningún primo.")
    return None

# =============================================================================
# EJECUCIÓN DE LA SEGUNDA PARTE
# =============================================================================
# Vamos a buscar el primer primo de tamaño N = 3 dígitos en nuestra cadena corta
print("--- INICIANDO BÚSQUEDA CON VENTANA DESLIZANTE ---")
buscar_primer_primo(cadena_e, tamano_bloque=3)
```

```python
from sympy import pi

def generar_decimales_pi_local(cantidad_digitos):
    """
    Genera la cantidad exacta de dígitos de pi directamente en la máquina local
    usando la librería de cálculo simbólico SymPy. Elimina la necesidad de internet.
    """
    print(f"Generando de forma matemática los primeros {cantidad_digitos} dígitos de pi...")

    # Evaluamos pi con la precisión deseada y lo convertimos a texto
    pi_completo_texto = str(pi.evalf(cantidad_digitos))

    # Filtramos para quedarnos únicamente con los dígitos numéricos
    cadena_numerica = "".join(caracter for caracter in pi_completo_texto if caracter.isdigit())

    # Como arranca en 314159..., le quitamos la parte entera (el 3)
    # para quedarnos puramente con la secuencia de decimales (14159...)
    decimales_pi = cadena_numerica[1:]

    return decimales_pi

# =============================================================================
# ACOPLAMIENTO FINAL DEL SISTEMA PARA PI (CÁLCULO SEGURO)
# =============================================================================
print(r"--- EJECUCIÓN FINAL DEL TRABAJO INTEGRAL ($\pi$) ---")

# 1. Generamos de forma local 2000 dígitos (suficientes para hallar el primo de 10)
CANTIDAD_DIGITOS = 2000
decimales_reales_pi = generar_decimales_pi_local(CANTIDAD_DIGITOS)

if decimales_reales_pi:
    print(f"¡Generación exitosa! Se procesará una secuencia de {len(decimales_reales_pi)} decimales.")

    # 2. Ejecutamos la Ventana Deslizante buscando un bloque de N = 10 dígitos
    N = 10
    buscar_primer_primo(decimales_reales_pi, tamano_bloque=N)
else:
    print("No se pudo iniciar el análisis debido a un error en la generación de datos.")
```
